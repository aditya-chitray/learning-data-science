#### 4. Strings

```python
# check if one string is substring of another string
my_str = "Hello World!"

print("Hello" in my_str)
print("hey" in my_str)

#check length of a string
print(len(my_str))

# Strings are immutable: i.e. parts of string cannot be changed
```

```python
# String concatenation

str1 = 'Hello'
str2 = 'World'

new_str = str1 + ' ' + str2
print(new_str)
```

```python
# String Interpolation

num1 = 5
num2 = 10

print(f'The sum of {num1} and {num2} is equal to {num1+num2}') 
```

```python
# String Slicing

print(my_str[0])
print(my_str[0:4]) # upper limit is not inclusive
print(my_str[:4])
print(my_str[6:]) 

#print(str[start:stop:step])

### reverse a string
print(my_str[::-1])
```

```python
# Strting Methods

print(my_str.upper())
print(my_str.lower())
print(my_str.split())
my_list = ['hello', 'world']
joined_my_str = ' '.join(my_list)
print(joined_my_str)
print(my_str.startswith('Hello'))
print(my_str.endswith('!'))

# other string methods: find, count, capitalize, isupper, islower, title
```

```python
## deleting items from list

developer = ['Jane Doe', 23, 'Python Developer']
del developer[1]
print(developer) # ['Jane Doe', 'Python Developer']
```

```python
### (in) is used to find if an element is part of list

programming_languages = ['Python', 'Java', 'C++', 'Rust']

print('Rust' in programming_languages) # True
print('JavaScript' in programming_languages) # False
```

```python
### Nested List

developer = ['Alice', 25, ['Python','C++','Rust']]
print(developer[2]) # ['Python', 'C++', 'Rust']
print(developer[2][1]) # C++
```

```python
## Unpacking values from list

developer = ['Alice', 34, 'Rust Developer']
name, age, job = developer

print(name) # 'Alice'
print(age) # 34
print(job) # 'Rust Developer'

developer = ['Alice', 34, 'Rust Developer']
name, *rest = developer

print(name) # 'Alice'
print(rest) # [34, 'Rust Developer']
```

```python
## Special ways of slicing
numbers = [1, 2, 3, 4, 5, 6]
numbers[1::2] # [2, 4, 6]
```

```python
### append(), sort(), pop()

numbers = [1,2,3,4,5]
numbers.append(6)
print(numbers)

numbers.pop()
print(numbers)

numbers = [3,2,1]
numbers.sort()
print(numbers)
```

```python
numbers = [1,2,3,4,5]
num2= [9,10,11]
numbers.append(num2) 
print(numbers) # [1, 2, 3, 4, 5, [9, 10, 11]]

## joining 2 list

num1 = [1,2,3]
num2 = [4,5,6]
num1.extend(num2)
print(num1) # [1, 2, 3, 4, 5, 6]
```

> insert(index,value): insert value at a specific index 
> remove(value): remove a specific value from list
> pop(index): remove value from a specific index
> clear(): clear the list of all values
> reverse(): reverse the order of elements of list
> index(): gives index of particular element of the list

```python
## zip ()

developers = ['Naomi', 'Dario', 'Jessica', 'Tom']
ids = [1, 2, 3, 4]

for name, id in zip(developers, ids):
    print(f'Name: {name}')
    print(f'ID: {id}')
```
