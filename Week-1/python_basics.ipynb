#### 2. User Input

```python
# basic user input
name = input("Please enter your name: ")
print("Hello, " + name)
```

```python

# taking input of other data types
age = int(input("Please provide your age: "))
print(f"You are {age} years old")

salary = float(input("Please enter your annual salary; "))
monthly_salary = salary/12
print(f"Your monthly salary is {monthly_salary} rupees")
```

```python

### Complex Inputs ###

x, y = map(int, input("Enter two numbers: ").split())
print(x + y)
```

```python

### Input with Validation ###

try:
    marks = float(input("Please enter your marks: "))
    print("Your marks: ", marks)

except ValueError:
    print("Invalid input. Please enter a number between 0-99")

```

```python

### Providing a list of data ###

prices = list(map(float,input("Enter prices separated by spaces: ").split()))
average_price = sum(prices)/len(prices)
print("Average Price: ",average_price)
```

#### Exercise 1: Take a number as input and print whether the number is positive ,negative or zero

```python
try:
    num = float(input("Please enter an number: "))
    if num >0:
        print("Positive")
    elif num <0:
        print("Negative")
    else:
        print("zero")
except ValueError:
    print("Invalid input! Please provide an number")
```

#### Exercise 2: Print all even numbers between 1 and 50

```python
for i in range(1,51):
    if i%2==0:
        print(i)
    else:
        continue
```

#### 4. Strings

```python
# check if one string is substring of another string
my_str = "Hello World!"

print("Hello" in my_str)
print("hey" in my_str)

#check length of a string
print(len(my_str))

# Strings are immutable: i.e. parts of string cannot be changed
```

```python
# String concatenation

str1 = 'Hello'
str2 = 'World'

new_str = str1 + ' ' + str2
print(new_str)
```

```python
# String Interpolation

num1 = 5
num2 = 10

print(f'The sum of {num1} and {num2} is equal to {num1+num2}') 
```

```python
# String Slicing

print(my_str[0])
print(my_str[0:4]) # upper limit is not inclusive
print(my_str[:4])
print(my_str[6:]) 

#print(str[start:stop:step])

### reverse a string
print(my_str[::-1])
```

```python
# Strting Methods

print(my_str.upper())
print(my_str.lower())
print(my_str.split())
my_list = ['hello', 'world']
joined_my_str = ' '.join(my_list)
print(joined_my_str)
print(my_str.startswith('Hello'))
print(my_str.endswith('!'))

# other string methods: find, count, capitalize, isupper, islower, title
```
