#### 2. Looping over Dictionary

```python
products= {
    'Laptop': 990,
    'Smartphone': 600,
    'Tablet': 250,
    'Headphone': 70
}

# # loop over values

# for price in products.values():
#     print(price)

# # loop over keys

# for product in products.keys():
#     print(product)

# #or 

# for product in products:
#     print(product)

# loop over both keys and values

# for product, price in products.items():
#     print(product, price)

# Let's say we want to reduce prices by 20% as part of discount. To calculate new prices

for product,price in products.items():
    products[product] = round(price*0.8)

print(products)
```

```python
for product in enumerate(products):
    print(product)

for index, product in enumerate(products.items()):
    print(index, product)
```
