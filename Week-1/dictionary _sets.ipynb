#### 1. Dictionary

```python
# define a dictionary

pizza = dict([('name','Margherita Pizza'),('price', 8.9),('calories_per_slice',250),('toppings', ['mozzarella', 'basil'])])

print(pizza.keys())
print(pizza.values())

pizza.get('name',[])
```

```python
## view object

pizza.items()

## clear all entries in dictionary

pizza.clear()
print(pizza)
```
